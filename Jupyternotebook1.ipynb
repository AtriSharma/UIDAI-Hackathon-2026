```python
import pandas as pd
import re

# Step 1: Clean the district column
Biometric['district_clean'] = Biometric['district'].astype(str).str.strip().str.lower()

# Remove symbols (*, ?, ., etc.)
Biometric['district_clean'] = Biometric['district_clean'].apply(lambda x: re.sub(r'[^\w\s\-&]', '', x))

# Replace multiple spaces with single space
Biometric['district_clean'] = Biometric['district_clean'].str.replace(r'\s+', ' ', regex=True)

```

```python
# Step 2: Mapping of wrong/variant names -> correct district names
district_map = {
    'mahabub nagar': 'Mahabubnagar',
    'medchal-malkajgiri': 'Medchal–Malkajgiri',
    'medchal?malkajgiri': 'Medchal–Malkajgiri',
    'rangareddi': 'Ranga Reddy',
    'k.v.rangareddy': 'Ranga Reddy',
    'mumbai sub urban': 'Mumbai Suburban',
    'mumbai( sub urban )': 'Mumbai Suburban',
    'south 24 pargana': 'South 24 Parganas',
    'purba medinipur': 'Purba Medinipur',
    'west midnapore': 'Paschim Medinipur',
    'hooghly': 'Hooghly',
    'haora': 'Howrah',
    'shimoga': 'Shivamogga',
    'bengaluru south': 'Bengaluru South',
    'bengaluru rural': 'Bengaluru Rural',
    'chikkamagaluru': 'Chikkamagaluru',
    'tiruvallur': 'Tiruvallur',
    'tiruvannamalai': 'Tiruvannamalai',
    'tiruppur': 'Tiruppur',
    'dadra & nagar haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'dadra and nagar haveli': 'Dadra and Nagar Haveli and Daman and Diu',
}
```

```python
# Step 3: Apply mapping, leave unmatched districts as-is
Biometric['district_clean'] = Biometric['district_clean'].apply(lambda x: district_map.get(x, x))
```

```python
# Step 4: Convert all districts to Title Case
Biometric['district_clean'] = Biometric['district_clean'].str.title()
```

```python
# Step 5: Replace original column with cleaned column
Biometric['district'] = Biometric['district_clean']
Biometric.drop(columns=['district_clean'], inplace=True)
```

```python
# Step 6: Save cleaned CSV
Biometric.to_csv('Biometric.csv', index=False)
```
